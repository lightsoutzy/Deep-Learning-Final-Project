```python
import numpy as np
import cv2
import os
```

```python
os.listdir("./Tea_Leaf_Disease/")
```

```python
new_data_path = "./Dataset/"
#List all sub directories in the main dataset
for d in os.listdir("./Tea_Leaf_Disease/"):
    if d[0]!='.':
        count = 1
        #List all images in each class 
        for image in os.listdir(os.path.join("./Tea_Leaf_Disease/",d)):
            print(os.path.join("./Tea_Leaf_Disease/",d,image))
            #Reading each image of each class
            img = cv2.imread(os.path.join("./Tea_Leaf_Disease/",d,image))
            #Checking if the shape is 256x256x3
            if img.shape != (256,256,3):
                #Resize the images to 256x256x3 if it is not already in that shape
                img = cv2.resize(img,(256,256),interpolation=cv2.INTER_CUBIC)
            #Define the new path to save the data 
            filename = new_data_path+d+"/image_"+str(count)+".png"
            #Save the resized images
            cv2.imwrite(filename,img)
            count+=1
```
